```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Set seed for reproducibility
np.random.seed(42)
n = 100

# Generate synthetic data for threat detection and response KPIs
data = pd.DataFrame({
    'Mean_Time_to_Detect': np.random.uniform(12, 48, n),  # MTTD in hours
    'Mean_Time_to_Acknowledge': np.random.uniform(1, 6, n),  # MTTA in hours
    'Mean_Time_to_Contain': np.random.uniform(4, 24, n),  # MTTC in hours
    'Mean_Time_to_Resolve': np.random.uniform(24, 120, n),  # MTTR in hours
    'Virus_Infection_Scan_Frequency': np.random.randint(1, 10, n),  # Scans per day
    'Malware_Detection_Rate': np.random.uniform(80, 99, n),  # Detection rate in %
    'Response_and_Recovery_Cost': np.random.uniform(1e6, 10e6, n)  # Cost in $ (1-10M)
})

# Display the first few rows
print(data.head())
```

```python
import matplotlib.pyplot as plt

# Scatter Plot: Mean Time to Resolve vs. Response and Recovery Cost
plt.figure(figsize=(10, 6))
plt.scatter(data['Mean_Time_to_Resolve'], data['Response_and_Recovery_Cost'], color='purple', alpha=0.7, edgecolor='k')
plt.title('Mean Time to Resolve vs. Response and Recovery Cost', fontsize=14)
plt.xlabel('Mean Time to Resolve (hours)', fontsize=12)
plt.ylabel('Response and Recovery Cost ($)', fontsize=12)
plt.ticklabel_format(style='plain', axis='y')  # Ensures plain formatting for y-axis
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Line Plot for Mean Time to Detect vs. Mean Time to Resolve
plt.figure(figsize=(10, 6))
plt.plot(data['Mean_Time_to_Detect'], label='Mean Time to Detect (MTTD)', linestyle='--', marker='o', color='blue')
plt.plot(data['Mean_Time_to_Resolve'], label='Mean Time to Resolve (MTTR)', linestyle='-', marker='x', color='red')
plt.title('Comparison of Mean Time to Detect and Mean Time to Resolve', fontsize=14)
plt.xlabel('Observations', fontsize=12)
plt.ylabel('Time (hours)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
