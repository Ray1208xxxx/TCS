```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define AI groups and tools
ai_groups = {
    "Generative AI Tools": ["ChatGPT", "CoPilot", "Bard (Google)"],
    "Business Process Automation Tools": ["UiPath", "ServiceNow", "Blue Prism"],
    "Customer Relationship Management (CRM) Tools": ["Salesforce Einstein", "HubSpot CRM", "Zoho CRM"]
}

# Generating mock data
data = []
np.random.seed(42)
for group, tools in ai_groups.items():
    for tool in tools:
        tool_data = {
            "Group": group,
            "AI Tool": tool,
            "Integration Cost ($)": np.random.randint(50000, 150000),
            "Annual Subscription Fee ($)": np.random.randint(20000, 50000),
            "Hours Saved per User per Month": np.random.randint(10, 50),
            "Pre-AI Errors": np.random.randint(200, 500),
            "Post-AI Errors": np.random.randint(50, 150),
            "Task Completion Time (Pre-AI, mins)": np.random.randint(60, 120),
            "Task Completion Time (Post-AI, mins)": np.random.randint(30, 60),
            "Pre-AI Employee Satisfaction Index": np.random.uniform(3.0, 4.5),
            "Post-AI Employee Satisfaction Index": np.random.uniform(4.0, 5.0),
            "Pre-AI Customer Satisfaction Score": np.random.uniform(70, 85),
            "Post-AI Customer Satisfaction Score": np.random.uniform(85, 95),
            "Average Cost per Error ($)": np.random.randint(1000, 5000),
            "Time to Full Adoption (weeks)": np.random.randint(4, 12),
            "Employee Hourly Wage ($)": np.random.randint(30, 50)
        }
        data.append(tool_data)

df = pd.DataFrame(data)

# Calculate metrics
df["Error Reduction Ratio (ERR)"] = (df["Pre-AI Errors"] - df["Post-AI Errors"]) / df["Pre-AI Errors"]
df["Employee Satisfaction Gain (ESG)"] = df["Post-AI Employee Satisfaction Index"] - df["Pre-AI Employee Satisfaction Index"]
df["Adoption Velocity (AV)"] = 100 / df["Time to Full Adoption (weeks)"]
df["Customer Experience Improvement (CXI)"] = (df["Post-AI Customer Satisfaction Score"] - df["Pre-AI Customer Satisfaction Score"]) / df["Pre-AI Customer Satisfaction Score"]
df["Cost Efficiency Index (CEI)"] = (
    df["Hours Saved per User per Month"] * df["Employee Hourly Wage ($)"] * 12
) / (df["Integration Cost ($)"] + df["Annual Subscription Fee ($)"])

metric_columns = ["Error Reduction Ratio (ERR)", "Employee Satisfaction Gain (ESG)",
                  "Adoption Velocity (AV)", "Customer Experience Improvement (CXI)",
                  "Cost Efficiency Index (CEI)"]

# Visualizations
for metric in metric_columns:
    plt.figure(figsize=(12, 6))
    for group in ai_groups.keys():
        group_data = df[df["Group"] == group]
        plt.bar(group_data["AI Tool"], group_data[metric], label=group)

    plt.title(f"{metric} Across All Groups", fontsize=16)
    plt.xlabel("AI Tool", fontsize=14)
    plt.ylabel(metric, fontsize=14)
    plt.xticks(rotation=45, fontsize=12)
    plt.legend(title="Groups", fontsize=12)
    plt.tight_layout()
    plt.show()

plt.figure(figsize=(12, 8))
sns.heatmap(df[metric_columns].set_index(df["AI Tool"]), annot=True, fmt=".2f", cmap="YlGnBu")
plt.title("Heatmap of Key Metrics Across AI Tools", fontsize=16)
plt.ylabel("AI Tool", fontsize=12)
plt.xlabel("Metrics", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```
